```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([-7., -4., -1., 2., 5., 8., 11., 15.])
y = np.array([3., 6., 9., 12., 15., 18., 21., 24.])
X, y

X =tf.cast(tf.constant(X), dtype=tf.float32 )
y = tf.cast(tf.constant(y), dtype=tf.float32)
X,y
# plt.scatter(X, y)
input_shape = X[0].shape
output_shape =y[0].shape
input_shape, output_shape
```

```python
# Basic model

import numpy as np
import tensorflow as tf

X = np.reshape(X, (-1, 1))

# Creating the model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1))

# compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

# fit the model
model.fit(X, y, epochs=5)
```

```python
# making prediction
x=[17.0]
y_pred = model.predict(x)
y_pred
# ploting our model
def plot_prediction(training_data= X,trainiing_label= y,testing_data= x,testing_label= 27 ,prediction= y_pred):
  plt.figure(figsize=(10,7))
  plt.scatter(training_data,trainiing_label, c='b', label= 'training data')
  plt.scatter(testing_data,testing_label, c='y', label= 'testing data')
  # plotting model predictions
  plt.scatter(testing_data, prediction, c='r', label="prediction" , marker=".")
  plt.legend()

plot_prediction()
```

```python
# Improving the model
import numpy as np
import tensorflow as tf

X = np.reshape(X, (-1, 1))

# Creating the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation ='sigmoid'),
    tf.keras.layers.Dense(100, activation ='sigmoid'),
    tf.keras.layers.Dense(100, activation ='relu'),
    tf.keras.layers.Dense(1)
])


# compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=.01))

# fit the model
model.fit(X, y, epochs=100)
```

```python
y_pred = model.predict([17.0])
y_pred
```

```python
# creating and training a new model
X = tf.range(-100, 100, 4)
y = X + 10

X = tf.reshape(X, (-1, 1))
X_train = X[:40]
y_train = y[:40]
X_test = X[40: ]
y_test = y[40: ]
# X_train, y_train, X_test, y_test
# creating my model

model_2= tf.keras.Sequential()
model_2.add(tf.keras.layers.Dense(100, activation="relu",))
model_2.add(tf.keras.layers.Dense(100, activation="relu"))
model_2.add(tf.keras.layers.Dense(1))

# compiling my model
model_2.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
              metrics=['mae'])

# fitting the model
model_2.fit(X_train, y_train, epochs=100)
```

```python
# checking the model
y_pred=model_2.predict(X_test)

y_pred, y_test
```

```python
# ploting our model
def plot_prediction(training_data= X_train,trainiing_label= y_train,testing_data= X_test,testing_label= y_test,prediction= y_pred):
  plt.figure(figsize=(10,7))
  plt.scatter(training_data,trainiing_label, c='b', label= 'training data')
  plt.scatter(testing_data,testing_label, c='y', label= 'testing data')
  # plotting model predictions
  plt.scatter(testing_data, prediction, c='r', label="prediction" , marker=".")
  plt.legend()

plot_prediction()
```
